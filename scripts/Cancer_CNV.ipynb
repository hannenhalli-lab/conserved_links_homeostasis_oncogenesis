```r
calculate_cnv_enrichment <- function(cnv_freq) {
    
    cnv_matrix <- numeric(length(combined_recurrent))
    for(i in 1:length(tcgaCNV)) {
        cnv_df <- na.omit(cnv_freq[[i]])

        ordered_freqs <- cnv_df[order(cnv_df, decreasing = T)]
        thresh <- min(ordered_freqs[1:((length(ordered_freqs))*.05)])
        
        back_thresh_freq <- cnv_df[cnv_df > thresh]
        
        exp_freq_vec <- numeric(length(combined_recurrent))
        
        for(j in 1:length(combined_recurrent)){
            current_set <- combined_recurrent[[j]]
            current_name <- names(combined_recurrent)[j]
            
            human_ensmbl[human_ensmbl$`Gene stable ID` %in% names(cnv_df)]
            tcgaCNV[[1]][rownames(tcgaCNV[[1]]) %in% human_ensmbl$`Gene stable ID`,]
            
            fore_thresh_freq_ensembl <- cnv_df[cnv_df > thresh]
            
            fore_thresh_freq_gn <- human_ensmbl[human_ensmbl$`Gene stable ID` %in% names(fore_thresh_freq_ensembl),]$`Gene name`
            current_foreground_gn <- human_ensmbl[human_ensmbl$`Gene stable ID` %in% names(cnv_df),]$`Gene name`
            
            current_foreground_total <- current_foreground_gn[current_foreground_gn %in% current_set]
            
            fore_thresh_freq <- fore_thresh_freq_gn[fore_thresh_freq_gn %in% current_set]
            
            exp_fisher_contingency = rbind(c(length(fore_thresh_freq),length(current_foreground_total)-length(fore_thresh_freq)),
                                       c(length(back_thresh_freq),length(cnv_df)-length(back_thresh_freq)))
            
            thresh_test <- fisher.test(exp_fisher_contingency)
        
#           print("significant and positive")
#           print(exp_fisher_contingency)
#           print(paste("Pvalue: ", fisher.test(exp_fisher_contingency)$p.value, sep=""))
#           print(paste("Odds: ", fisher.test(exp_fisher_contingency)$estimate, sep=""))
#           print("")
        
        
            if(thresh_test$p.value < .05) {
                exp_freq_vec[j] = thresh_test$estimate
            } else { exp_freq_vec[j] = -1}
        }

        cnv_matrix <- cbind(cnv_matrix, exp_freq_vec)

    }

    cnv_matrix <- cnv_matrix[,2:ncol(cnv_matrix)]
    colnames(cnv_matrix) <- names(tcgaCNV)
    rownames(cnv_matrix) <- names(combined_recurrent)
    return(t(cnv_matrix))
    
}
```

```r
amp_matrix <- calculate_cnv_enrichment(CNV_amp_freq)
del_matrix <- calculate_cnv_enrichment(CNV_del_freq)
```

```r
#current_list <- immediate_combined_recurrent

get_cnv_data <- function(current_list, matrix_order) {

    amp_mean_mat <- NULL
    amp_median_mat <- NULL
    del_mean_mat <- NULL
    del_median_mat <- NULL
    
    amp_frac_mat <- NULL
    del_frac_mat <- NULL

    amp_plot_list <- list()
    del_plot_list <- list()
    for(i in 1:length(tcgaCNV)) {
        current_cancer_name <- names(tcgaCNV)[i]
        current_df <- tcgaCNV[[current_cancer_name]]

        amplification_frac <- apply(current_df,1,FUN = function(x) {amp = sum(x>0)
                                                   return(amp/length(x))})
        amp_zscore = (amplification_frac - mean(amplification_frac)) / sd(amplification_frac)
        deletion_frac <- apply(current_df,1,FUN = function(x) {del = sum(x<0)
                                                   return(del/length(x))})
        del_zscore = (deletion_frac - mean(deletion_frac)) / sd(deletion_frac)

        amp_score_vec <- amp_zscore
        amp_name_vec <- rep("Total_zcores", length(amp_score_vec))

        del_score_vec <- del_zscore
        del_name_vec <- rep("Total_zcores", length(del_score_vec))

        amp_mean_vec <- numeric(length((current_list)))
        amp_median_vec <- numeric(length((current_list)))

        del_mean_vec <- numeric(length((current_list)))
        del_median_vec <- numeric(length((current_list)))
        
        amp_frac_vec <- numeric(length((current_list)))
        del_frac_vec <- numeric(length((current_list)))
        for(j in 1:length((current_list))) {

            current_hp_name <- names(current_list)[j]
            current_hp_vec <- current_list[[current_hp_name]]

            ensmbl_ref <- human_ensmbl[human_ensmbl$`Gene name` %in% current_hp_vec,]$`Gene stable ID`

            current_amp_vec <- amp_zscore[which(names(amp_zscore) %in% ensmbl_ref)]
            amp_mean_vec[j] <- mean(current_amp_vec)
            amp_median_vec[j] <- median(current_amp_vec)
            amp_score_vec <- append(amp_score_vec, current_amp_vec, length(amp_score_vec))
            amp_name_vec <- append(amp_name_vec, rep(current_hp_name, length(current_amp_vec)),
                                   length(amp_name_vec))

            current_del_vec <- del_zscore[which(names(del_zscore) %in% ensmbl_ref)]
            del_median_vec[j] <- median(current_del_vec)
            del_mean_vec[j] <- mean(current_del_vec)

            del_score_vec <- append(del_score_vec, current_del_vec, length(del_score_vec))
            del_name_vec <- append(del_name_vec, rep(current_hp_name,
                                                     length(current_del_vec)),
                                   length(del_name_vec))
            
            
            ### fractional data
            current_amp_frac_vec <- amplification_frac[which(names(amplification_frac) %in% ensmbl_ref)]
            top_preturbed_amp <- current_amp_frac_vec[order(current_amp_frac_vec, decreasing=T)][1:50]
            amp_frac_vec[j] <- mean(top_preturbed_amp)
            
            current_del_frac_vec <- deletion_frac[which(names(deletion_frac) %in% ensmbl_ref)]
            top_preturbed_del <- current_del_frac_vec[order(current_del_frac_vec, decreasing=T)][1:50]
            del_frac_vec[j] <- mean(top_preturbed_del)

        }                       
        if(i == 1) {
            amp_mean_mat <- amp_mean_vec
            amp_median_mat <- amp_median_vec
            del_mean_mat <- del_mean_vec
            del_median_mat <- del_median_vec
            
            amp_frac_mat <- amp_frac_vec
            del_frac_mat <- del_frac_vec
            
        } else {
            amp_mean_mat <- cbind(amp_mean_mat, amp_mean_vec)
            amp_median_mat <- cbind(amp_median_mat, amp_median_vec)
            del_mean_mat <- cbind(del_mean_mat, del_mean_vec)
            del_median_mat <- cbind(del_median_mat, del_median_vec)
            
            
            amp_frac_mat <- cbind(amp_frac_mat, amp_frac_vec)
            del_frac_mat <- cbind(del_frac_mat, del_frac_vec)
            
        }

        amp_df <- data.frame("Zscores" = amp_score_vec,
                             "Label" = amp_name_vec)
        del_df <- data.frame("Zscores" = del_score_vec,
                             "Label" = del_name_vec)


        options(repr.plot.width=10, repr.plot.height=10)
        amp_plot_list[[current_cancer_name]] <- ggplot(amp_df, aes(x=Label, y=Zscores,
                                       fill=Label)) +
              geom_boxplot() +
              ggtitle(paste("Plot of", current_cancer_name, "Amplification Zscores", sep =" "))  + 
              theme(axis.text.x = element_text(angle = 90))

        del_plot_list[[current_cancer_name]] <- ggplot(del_df, aes(x=Label, y=Zscores,
                                       fill=Label)) +
              geom_boxplot() +
              ggtitle(paste("Plot of", current_cancer_name, "Deletion Zscores", sep =" "))  + 
              theme(axis.text.x = element_text(angle = 90))
    }
    
    colnames(amp_mean_mat) <- names(tcgaCNV)
    rownames(amp_mean_mat) <- names(current_list)
    
    colnames(amp_median_mat) <- names(tcgaCNV)
    rownames(amp_median_mat) <- names(current_list)
    
    colnames(del_mean_mat) <- names(tcgaCNV)
    rownames(del_mean_mat) <- names(current_list)
    
    colnames(del_median_mat) <- names(tcgaCNV)
    rownames(del_median_mat) <- names(current_list)
    
    colnames(amp_frac_mat) <- names(tcgaCNV)
    rownames(amp_frac_mat) <- names(current_list)
    
    colnames(del_frac_mat) <- names(tcgaCNV)
    rownames(del_frac_mat) <- names(current_list)
    
    
    
    amp_plot_list[["amp_mat_mean"]] <- pheatmap(amp_mean_mat[matrix_order,],
                                                color = colorRampPalette(c("white", "white", "white", "white",'lightblue1',"lightblue2", "lightblue3", "lightblue4","dodgerblue3", "dodgerblue4"))(10),
                                                fontsize = 14,
                                                breaks = c(-2, -1.5, -1, -.5, 0, .5, 1, 1.5, 2, 2.5),
                                                cluster_rows=F, cluster_cols=F)
    
    amp_plot_list[["amp_mat_median"]] <- pheatmap(amp_median_mat[matrix_order,],
                                                  color = colorRampPalette(c("white", "white", "white", "white",'lightblue1',"lightblue2", "lightblue3", "lightblue4","dodgerblue3", "dodgerblue4"))(10),
                                                  fontsize = 14,
                                                  breaks = c(-2, -1.5, -1, -.5, 0, .5, 1, 1.5, 2, 2.5),
                                                  cluster_rows=F, cluster_cols=F)
    amp_plot_list[["amp_mat_top_frac"]] <- pheatmap(amp_frac_mat[matrix_order,],
                                                  color = colorRampPalette(c("white",'lightblue1',"lightblue2", "lightblue3", "lightblue4"))(10),
                                                  fontsize = 14,
                                                  breaks = seq(0,1,.1),
                                                  cluster_rows=F, cluster_cols=F)
    
    del_plot_list[["del_mat_mean"]] <- pheatmap(del_mean_mat[matrix_order,],
                                                color = colorRampPalette(c("white", "white", "white", "white",'lightblue1',"lightblue2", "lightblue3", "lightblue4","dodgerblue3", "dodgerblue4"))(10),
                                                fontsize = 14,
                                                breaks = c(-2, -1.5, -1, -.5, 0, .5, 1, 1.5, 2, 2.5),
                                                cluster_rows=F, cluster_cols=F)
    
    del_plot_list[["del_mat_median"]] <- pheatmap(del_median_mat[matrix_order,],
                                                  color = colorRampPalette(c("white", "white", "white", "white",'lightblue1',"lightblue2", "lightblue3", "lightblue4","dodgerblue3", "dodgerblue4"))(10),
                                                  fontsize = 14,
                                                  breaks = c(-2, -1.5, -1, -.5, 0, .5, 1, 1.5, 2, 2.5),
                                                  cluster_rows=F, cluster_cols=F)
    del_plot_list[["del_mat_top_frac"]] <- pheatmap(del_frac_mat[matrix_order,],
                                                  color = colorRampPalette(c("white",'lightblue1',"lightblue2", "lightblue3", "lightblue4"))(10),
                                                  fontsize = 14,
                                                  breaks = seq(0,1,.1),
                                                  cluster_rows=F, cluster_cols=F)
    
    return(list("amp" = amp_plot_list,
                "del" = del_plot_list))
}
                               
```
