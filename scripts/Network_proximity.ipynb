```r
hPPIN_mean_sp <- data.frame(fread("../ppi_analysis/extra_files/hPPIN_meanSP_HP_Phenotypes_hallmark_9by9.csv"))
rownames(hPPIN_mean_sp) <- hPPIN_mean_sp$V1
hPPIN_mean_sp_9x9 <- hPPIN_mean_sp[,-c(1)]

hPPIN_matched_mean_sp <- data.frame(fread("../ppi_analysis/extra_files/hPPIN_meanSP_HPmatched_Phenotypes_hallmark_9by9.csv"))
rownames(hPPIN_matched_mean_sp) <- hPPIN_matched_mean_sp$V1
hPPIN_matched_mean_sp_9x9 <- hPPIN_matched_mean_sp[,-c(1)]

hPPIN_pvals <- data.frame(fread("../ppi_analysis/extra_files/hPPIN_matched_pvals_hallmark_9by9.csv"))
rownames(hPPIN_pvals) <- hPPIN_pvals$V1
hPPIN_pvals_9x9 <- hPPIN_pvals[,-c(1)]

hPPIN_LFC <- data.frame(fread("../ppi_analysis/extra_files/hPPIN_matched_LFC_hallmark_9by9.csv"))
rownames(hPPIN_LFC) <- hPPIN_LFC$V1
hPPIN_LFC_9x9 <- hPPIN_LFC[,-c(1)]
```

```r
options(repr.plot.width=14, repr.plot.height=8)
pval_file <- hPPIN_pvals_9x9
lfc_file <- hPPIN_LFC_9x9
new_plot <- matrix(nrow = nrow(pval_file), ncol = ncol(pval_file))
for(i in 1:nrow(pval_file)) {
    for(j in 1:length(pval_file)){
        if(pval_file[i,j] < 0.5) {
            new_plot[i,j] <- lfc_file[i,j]
        }
        else {
            new_plot[i,j] <- range(lfc_file)[1] - .1
        }
    }
}
rownames(new_plot) <- rownames(pval_file)
colnames(new_plot) <- colnames(pval_file)

pheatmap(data.frame(new_plot),
         cluster_rows=FALSE,
         cluster_cols=FALSE,
         color = colorRampPalette(c("white","lightblue3","lightblue2","lightblue1","firebrick"))(5),
         fontsize = 14,
         breaks = c(range(lfc_file)[1] - .1, range(lfc_file)[1] - .00001, -.1,-.05,0, .05))
```

```r
library(pheatmap)
library(grid)
library(ggplot2)

pval_file <- hPPIN_pvals
lfc_file <- hPPIN_LFC

# Create matrix for plotting
new_plot <- matrix(nrow = nrow(pval_file), ncol = ncol(pval_file))
for(i in 1:nrow(pval_file)) {
    for(j in 1:ncol(pval_file)){
        if(pval_file[i,j] < 0.5) {
            new_plot[i,j] <- lfc_file[i,j]
        } else {
            new_plot[i,j] <- range(lfc_file)[1] - 0.1
        }
    }
}
rownames(new_plot) <- rownames(pval_file)
colnames(new_plot) <- colnames(pval_file)

# Capture the plot
p <- grid::grid.grabExpr(
  pheatmap(data.frame(new_plot),
           cluster_rows = FALSE,
           cluster_cols = FALSE,
           color = colorRampPalette(c("white", "lightblue3", "lightblue2", "lightblue1", "firebrick"))(5),
           fontsize = 15,
           breaks = c(range(lfc_file)[1] - 0.1, range(lfc_file)[1] - 0.00001, -0.33, -0.16, 0, 0.1))
)

# Save to PDF
ggsave("~/Desktop/hPPIN_heatmap.pdf", plot = p, width = 15, height = 10, dpi = 600, units = "in")
```
