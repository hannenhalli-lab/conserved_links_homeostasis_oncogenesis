```r
library("wesanderson")
```

```r
frac_files <- list.files("../data/", "gs.*Frac.RDS")
gs_ratios <- list()
gs_labels <- list()
for(i in 1:length(frac_files)) {
    curr_frac_gs <- readRDS(paste0("../data/",frac_files[i]))
    name <- gsub("(gs_high_|\\Frac.RDS)","",frac_files[i])
    gs_ratios[[name]] <- curr_frac_gs
    gs_labels[[name]] <- rep(name,length(curr_frac_gs))
    
}

frac_files_high <- list.files("../data/", "high_low.*Frac.RDS")
high_ratios <- list()
high_labels <- list()
for(i in 1:length(frac_files_high)) {
    curr_frac_high <- readRDS(paste0("../data/",frac_files_high[i]))
    name <- gsub("(high_low_|\\Frac.RDS)","",frac_files_high[i])
    high_ratios[[name]] <- curr_frac_high
    high_labels[[name]] <- rep(name,length(curr_frac_high))
    
}
```

```r
options(repr.plot.width=18, repr.plot.height=14)
```

```r
gs_df <- data.frame("label"= unlist(gs_labels),
                    "ratio" = 1/unlist(gs_ratios))
ggplot(gs_df, aes(y=ratio,x=label,fill= label)) +
geom_boxplot()  +
scale_y_continuous(limits=c(0,7)) +
geom_hline(aes(yintercept=1), colour="red") +
theme_minimal()+
scale_fill_manual(values = wes_palette("FantasticFox1",
                                       n = 7,type = "continuous"),
                  name = "Categories")+ 
theme(text = element_text(size=28),
          legend.position="none") +
scale_x_discrete(labels=c("DEG",
                          "DEG-up",
                          "DEG-down",
                          "DepMap",
                          "HR-Pos",
                          "HR-Neg",
                          "Mutation Frequency"))+
labs(title="UNITe-High vs. Goldset across modalities",x="", y = "Ratio of UNITe-High vs. Goldset across Cancer Types")

# add another boxplot for HR protective and hazardous
# flip
```
