```python
def list_contains(List1, List2): 
    check = []
  
    # add elements in list 1 that are in list2
    for m in List1: 
        if (m in List2) & (not (m in check)): 
            check.append(m)
            
    return check
```

```python
def submit_fractions_jobs(network_filename, network):
    
    files = os.listdir("/data/timonaj/cancer_as_wound/ppi_analysis/total_spec_exp_genes/")
    # write out matched control list
    for i in range(len(files)):
        filename = files[i]
        gs_name = filename[:-4]

        file_in1 = network_filename
        file_in2 = "/data/timonaj/cancer_as_wound/ppi_analysis/total_spec_exp_genes/" + filename
        file_in3 = gs_name 
        file_out = "/data/timonaj/cancer_as_wound/ppi_analysis/fractions_data/"
        file_out2 = network

        command = "printf \"#!/bin/bash\nmodule load python\npython /data/timonaj/cancer_as_wound/ppi_analysis/compute_fractions.py " + file_in1 + ' ' + file_in2 + ' ' + file_in3 + ' ' + file_out + ' ' + file_out2 + "\" > " + "./fractions_data/" + filename + ".job \nsbatch --mem=50g --cpus-per-task=4 --gres=lscratch:2 --time 8:00:00 " + "./fractions_data/" + filename + ".job \n"
        print(command)
        os.system(command)
```
